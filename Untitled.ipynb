## 문제 1) pandas 데이터프레임에 학습데이터셋 로드 하여, 결측치 제거 해주세요.

### 데이터 불러오기

```python
df = pd.read_csv('dt.csv', encoding='cp949')
df
```

### 데이터 정보 확인

```python
print(df.info())

print("\n################ 결측치 개수 ################")
print((df == "").sum())
```

```python
df[df.select_dtypes(object).columns.values] = df[df.select_dtypes(object).columns.values].fillna("")
df[df.select_dtypes(float).columns.values] = df[df.select_dtypes(float).columns.values].fillna(0)
df
```

## 문제 3) object 타입의 범주형 데이터를 수치형 데이터로 변환 하세요.

```python
df[df.select_dtypes(object).columns.values]
```

```python
from sklearn.preprocessing import LabelEncoder

encoder_dic = {}

for col in df[df.select_dtypes(object).columns.values]:
    encoder_dic[col] = LabelEncoder()
    encoder_dic[col].fit(df[col])
    df[col] = encoder_dic[col].transform(df[col])
    
df
```

## 문제 4) 범주형 변수의 데이터를 원-핫 인코딩 처리 하세요.

```python
np.setdiff1d(df.select_dtypes(object).columns.values, np.array(['수주여부']))
```

```python
df = pd.get_dummies(df, columns=np.setdiff1d(df.select_dtypes(object).columns.values, np.array(['수주여부'])))
df
```

## 문제 5) 훈련 데이터셋과 검증 데이터셋을 7:3 으로 분리하세요.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()

X = df[[c for c in df.columns if c != '수주여부']]
y = df['수주여부']

X_scaled = minmax_scaler.fit_transform(X)

X_train, X_valid, y_train, y_valid = train_test_split(X_scaled, y , test_size=0.3, shuffle=True, stratify=y, random_state=34)
```
